# PREPROCESSING OF TEXT

## Stop Words Removal

```python
stop = set(stopwords.words('english'))
print(stop)
```

```python
def check_word_in_stopwords(word):
    if word in set(stopwords.words('english')):
        return "Available"
```

```python
check_word_in_stopwords("is")
```

```python
import re
temp =[]
snow = nltk.stem.SnowballStemmer('english')

for sentence in messages.values:
    sentence = sentence.lower()               
    cleaner = re.compile('<.*?>')
    sentence = re.sub(cleaner, ' ', sentence)        
    sentence = re.sub(r'[?|!|\'|"|#|:]',r'',sentence)
    sentence = re.sub(r'[.|,|)|(|\|/|_|\\|@|=]',r' ',sentence)        
    
    words = [snow.stem(word) for word in sentence.split() if word not in stopwords.words('english')]   
    temp.append(words)
```

```python
final_messages = temp.copy()
```

```python
print(messages[0])
```

```python
print(final_messages[0])
```

```python
sentc = []
for row in final_messages:
    sequ = ''
    for word in row:
        sequ = sequ + ' ' + word
    sentc.append(sequ)
```

```python
sentc[0]
```

## Calling out the Bag of Words

```python
cv = CountVectorizer(max_features=5000)
```

```python
cv.fit_transform(sentc)
```
